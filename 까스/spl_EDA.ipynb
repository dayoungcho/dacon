```python
use_spl = pd.read_csv('한국가스공사_용도별 월 공급량_20250630.csv', encoding='cp949')
```

```python
use_spl['splym'] = use_spl['연도'].astype('str') + '-' + use_spl['월'].astype('str')
use_spl
```

```python
use_spl_city = use_spl[use_spl['용도']=='도시가스']
use_spl_gen= use_spl[use_spl['용도']=='발전']
```

```python
use_spl_city_agg = use_spl_city.groupby('연도').sum().reset_index()
use_spl_gen_agg = use_spl_gen.groupby('연도').sum().reset_index()

use_spl_city_agg = use_spl_city_agg[use_spl_city_agg['연도'] != 2025]
use_spl_gen_agg = use_spl_gen_agg[use_spl_gen_agg['연도'] != 2025]
```

```python
use_spl_gen_agg.연도
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=use_spl_gen['splym'], y=use_spl_gen['공급량'],mode = 'lines',name='발전용가스'))
fig.add_trace(go.Scatter(x=use_spl_city['splym'], y=use_spl_city['공급량'],mode = 'lines',name='도시가스'))
fig
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(x=use_spl_gen['splym'], y=use_spl_gen['공급량'], name="발전용가스")
)

fig.add_trace(
    go.Bar(x=use_spl_city['splym'], y=use_spl_city['공급량'], name="도시가스")
)

fig.update_layout(
    barmode='stack'
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(x=use_spl_gen_agg['연도'], y=use_spl_gen_agg['공급량'], name="발전용가스")
)

fig.add_trace(
    go.Bar(x=use_spl_city_agg['연도'], y=use_spl_city_agg['공급량'], name="도시가스")
)

fig.update_layout(
    barmode='stack'
)

fig.show()
```
